```python
import pandas as pd
import numpy as np

def load_and_join_fossil_data():
    # Load the datasets
    fossil_list = pd.read_csv('data/fossil_list.csv')
    jurassic_data = pd.read_csv('data/jurrasic raw.csv')
    
    # Clean column names in jurassic data by removing 'attributes_' prefix
    jurassic_data.columns = [col.replace('attributes_', '') for col in jurassic_data.columns]
    
    # Perform the join based on PROVINCE and DISTRICT
    joined_data = pd.merge(
        fossil_list,
        jurassic_data,
        left_on=['PROVINCE', 'DISTRICT'],
        right_on=['PROVINCE', 'DISTRICT'],
        how='inner'
    )
    
    return joined_data

def analyze_joined_data(joined_data):
    # Basic analysis of the joined dataset
    analysis = {
        'total_matches': len(joined_data),
        'unique_provinces': joined_data['PROVINCE'].nunique(),
        'unique_districts': joined_data['DISTRICT'].nunique(),
        'fossil_groups': joined_data['F_GROUP'].value_counts().to_dict(),
        'sites_per_province': joined_data['PROVINCE'].value_counts().to_dict()
    }
    
    return analysis

def export_joined_data(joined_data, output_file='joined_fossil_data.csv'):
    # Select relevant columns for the output
    columns_to_export = [
        'FOSSIL_ID', 'SCI_NAME', 'COM_NAME', 'F_GROUP', 'F_TYPE',
        'PROVINCE', 'DISTRICT', 'SITE_ID', 'NAME_TH', 'NAME_ENG',
        'LOCALITY', 'FOS_DES_TH', 'PERIODFROM'
    ]
    
    # Filter columns that exist in the joined dataset
    existing_columns = [col for col in columns_to_export if col in joined_data.columns]
    
    # Export to CSV
    joined_data[existing_columns].to_csv(output_file, index=False, encoding='utf-8')

def main():
    try:
        # Load and join the data
        print("Loading and joining datasets...")
        joined_data = load_and_join_fossil_data()
        
        # Analyze the joined data
        print("\nAnalyzing joined data...")
        analysis = analyze_joined_data(joined_data)
        
        # Print analysis results
        print("\nJoin Results:")
        print(f"Total matched records: {analysis['total_matches']}")
        print(f"Number of unique provinces: {analysis['unique_provinces']}")
        print(f"Number of unique districts: {analysis['unique_districts']}")
        
        print("\nFossil Groups Distribution:")
        for group, count in analysis['fossil_groups'].items():
            print(f"- {group}: {count}")
        
        print("\nSites per Province:")
        for province, count in analysis['sites_per_province'].items():
            print(f"- {province}: {count}")
        
        # Export the joined data
        output_file = 'joined_fossil_data.csv'
        print(f"\nExporting joined data to {output_file}...")
        export_joined_data(joined_data, output_file)
        
        print("\nProcess completed successfully!")
        
        # Return the joined dataset for further analysis if needed
        return joined_data
        
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None

if __name__ == "__main__":
    joined_data = main()
```

```python
import pandas as pd
import numpy as np

def perform_multiple_merges():
    # โหลดข้อมูล
    print("กำลังโหลดข้อมูล...")
    fossil_list = pd.read_csv('data/fossil_list.csv')
    jurassic_data = pd.read_csv('data/jurrasic raw.csv')
    
    # ลบ prefix 'attributes_' จาก jurassic columns
    jurassic_data.columns = [col.replace('attributes_', '') for col in jurassic_data.columns]
    
    print("\nข้อมูลเริ่มต้น:")
    print(f"Fossil list: {len(fossil_list)} แถว")
    print(f"Jurassic data: {len(jurassic_data)} แถว")

    # 1. Merge แบบละเอียด (PROVINCE + DISTRICT + TAMBOM)
    detailed_merge = pd.merge(
        fossil_list,
        jurassic_data,
        left_on=['PROVINCE', 'DISTRICT', 'TAMBOM'],
        right_on=['PROVINCE', 'DISTRICT', 'TAMBON'],
        how='inner',
        suffixes=('_fossil', '_site')
    )
    
    # 2. Merge แบบกลาง (PROVINCE + DISTRICT)
    medium_merge = pd.merge(
        fossil_list,
        jurassic_data,
        on=['PROVINCE', 'DISTRICT'],
        how='inner',
        suffixes=('_fossil', '_site')
    )
    
    # 3. Merge แบบกว้าง (PROVINCE only)
    broad_merge = pd.merge(
        fossil_list,
        jurassic_data,
        on=['PROVINCE'],
        how='inner',
        suffixes=('_fossil', '_site')
    )
    
    # 4. หาข้อมูลที่ไม่ถูก merge
    unmatched = pd.merge(
        fossil_list,
        jurassic_data,
        on=['PROVINCE'],
        how='outer',
        indicator=True,
        suffixes=('_fossil', '_site')
    ).query('_merge != "both"')
    
    # แสดงผลการวิเคราะห์
    print("\nผลการ Merge:")
    print(f"1. Merge แบบละเอียด (PROVINCE + DISTRICT + TAMBOM): {len(detailed_merge)} แถว")
    print(f"2. Merge แบบกลาง (PROVINCE + DISTRICT): {len(medium_merge)} แถว")
    print(f"3. Merge แบบกว้าง (PROVINCE): {len(broad_merge)} แถว")
    print(f"4. ข้อมูลที่ไม่ถูก merge: {len(unmatched)} แถว")
    
    # วิเคราะห์การกระจายตัวของข้อมูล
    print("\nการกระจายตัวตามจังหวัด (top 5):")
    print("\nแบบละเอียด:")
    print(detailed_merge['PROVINCE'].value_counts().head())
    print("\nแบบกลาง:")
    print(medium_merge['PROVINCE'].value_counts().head())
    print("\nแบบกว้าง:")
    print(broad_merge['PROVINCE'].value_counts().head())
    
    # บันทึกผลลัพธ์
    detailed_merge.to_csv('merge_detailed.csv', index=False)
    medium_merge.to_csv('merge_medium.csv', index=False)
    broad_merge.to_csv('merge_broad.csv', index=False)
    unmatched.to_csv('unmatched_data.csv', index=False)
    
    # วิเคราะห์ข้อมูลที่หายไป
    print("\nวิเคราะห์ข้อมูลที่ไม่ถูก merge:")
    unmatched_fossil = unmatched[unmatched['FOSSIL_ID'].notna()]
    print(f"\nจำนวนซากดึกดำบรรพ์ที่ไม่ถูก merge: {len(unmatched_fossil)}")
    if len(unmatched_fossil) > 0:
        print("\nจังหวัดที่มีซากดึกดำบรรพ์แต่ไม่มีข้อมูลแหล่ง:")
        print(unmatched_fossil['PROVINCE'].value_counts().head())
    
    return {
        'detailed': detailed_merge,
        'medium': medium_merge,
        'broad': broad_merge,
        'unmatched': unmatched
    }

# รันฟังก์ชัน
merged_data = perform_multiple_merges()
```
